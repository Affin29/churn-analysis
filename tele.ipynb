```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
pf["gender_num"] = le.fit_transform(pf["gender"])
pf["Partner_num"] = le.fit_transform(pf["Partner"])
pf["Dependents_num"] = le.fit_transform(pf["Dependents"])
pf["Churn_num"] = le.fit_transform(pf["Churn"])
pf['customerID_num']=le.fit_transform(pf['customerID'])
pf['SeniorCitizen_num']=le.fit_transform(pf['SeniorCitizen'])
pf['Partner_num']=le.fit_transform(pf['Partner'])
pf['Dependents_num']=le.fit_transform(pf['Dependents'])
pf['PhoneService_num']=le.fit_transform(pf['PhoneService'])
pf['MultipleLines_num']=le.fit_transform(pf['MultipleLines'])
pf['InternetService_num']=le.fit_transform(pf['InternetService'])
pf['OnlineSecurity_num']=le.fit_transform(pf['OnlineSecurity'])
pf['Contract_num']=le.fit_transform(pf['Contract'])
pf['PaperlessBilling_num']=le.fit_transform(pf['PaperlessBilling'])
pf['PaymentMethod_num']=le.fit_transform(pf['PaymentMethod'])
pf['Churn_num']=le.fit_transform(pf['Churn'])
pf['OnlineBackup_num']=le.fit_transform(pf['OnlineBackup'])
pf['DeviceProtection_num']=le.fit_transform(pf['DeviceProtection'])
pf['TechSupport_num']=le.fit_transform(pf['TechSupport'])
pf
```
